7) Which month is the best time to sell avocado

Ans = February and May are the best time to sell avocado.

```python
df[['Year','Month','Day']] = df['Date'].str.split('-',expand = True)
df.head()
```

```python
df['Year_month'] = df['Year'] + df['Month']
df.head()
```

```python
df.groupby('Year_month')['Total Volume'].sum()
```

```python
df.groupby('Year_month')['Total Volume'].sum().plot.line(figsize=(15,4))
```

```python
df.groupby('Year_month')['Total Volume'].sum().plot.bar(figsize=(15,4))
```

From the graph of sum Total Volume of avocado, seem like May is the best time to sell avocado. \
We will test another way on the mean of Total Volume of avocado in each month.

```python
df.groupby('Month')['Total Volume'].mean()
```

```python
df.groupby('Month')['Total Volume'].mean().max()
```

```python
df.groupby('Month')['Total Volume'].mean().plot.bar(figsize=(5,3))
```

From above graph, February is the the best time to sell avocado but it dosen't make sense becuase year 2018 record only 3 months (Jan - Mar). \
So, I will drop dataframe of year 2018.

```python
index_names = df[df['year'] == 2018].index
df.drop(index_names, inplace=True)
```

```python
df.head()
```

```python
df.groupby('Month')['Total Volume'].mean()
```

```python
df.groupby('Month')['Total Volume'].mean().max()
```

```python
df.groupby('Month')['Total Volume'].mean().plot.bar(figsize=(5,3))
```

When drop year 2018 February has the highest mean of Total Volume, but February and May are sell equally well. \
So, February and May are the best time to sell avocado.
